## SVC (kernek="linear")

```python
from sklearn.svm import SVC
from sklearn import datasets

iris = datasets.load_iris()
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = iris["target"]

setosa_or_versicolor = (y == 0) | (y == 1)
X = X[setosa_or_versicolor]
y = y[setosa_or_versicolor]

# SVM Classifier model
svm_clf = SVC(kernel="linear",C=100,gamma=1000)
svm_clf.fit(X, y)
```

```python
def plotting(model):
    #these are the w0 and w1 coefficients 
    w=model.coef_.flatten()
    ##w0x0+x1w1+b=0 
    #x1=-w0x0/w1 -b/w1
    x0=np.linspace(1,5.5,100)
    x1=-w[0]/w[1]*x0 - model.intercept_[0]/w[1]
    bound_up=x1+1/w[1]
    bound_down=x1-1/w[1]

    # now, I want to plot them with the dataset
    plt.plot(X[:,0][y==0],X[:,1][y==0],".", label="setosa")
    plt.plot(X[:,0][y==1],X[:,1][y==1], ".",label="versicolor")
    plt.plot(x0,x1)
    plt.plot(x0,bound_up,"--")
    plt.plot(x0,bound_down,"--")
    plt.legend()
```

```python
plotting(svm_clf)
```

## LinearSVC

```python
from sklearn.svm import LinearSVC


linsvc_clf=LinearSVC(C=5,loss="hinge")
linsvc_clf.fit(X,y)
```

```python
fig, axes = plt.subplots(ncols=2, figsize=(10,2.7), sharey=True)
                 
plt.sca(axes[0])
plotting(linsvc_clf)
plt.title("linsvc_clf")
plt.sca(axes[1])
plotting(svm_clf)
plt.title("svm_clf")
```

# SGD Classifier

```python
from sklearn.linear_model import SGDClassifier
C=0.5
sgd_clf = SGDClassifier(loss="hinge", max_iter=1000, tol=1e-2, random_state=42, alpha=1/(C*len(X)))
sgd_clf.fit(X, y)
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(20,5), sharey=True)
                 
plt.sca(axes[0])
plotting(linsvc_clf)
plt.title("linsvc_clf")
plt.sca(axes[1])
plotting(svm_clf)
plt.title("svm_clf")
plt.sca(axes[2])
plotting(sgd_clf)
plt.title("sgd_clf")
plt.plot(X[-2,0],X[-2,1],"o",label="wrong") 
plt.legend()
```

## expore different metrics

```python
from sklearn.model_selection import cross_val_predict
y_predicted=cross_val_predict(sgd_clf, X,y,cv=10)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y,y_predicted)
```
